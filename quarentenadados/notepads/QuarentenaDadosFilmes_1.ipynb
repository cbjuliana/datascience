```
import pandas as pd
```

```
#função do pandas para ler um arquivo csv
filmes = pd.read_csv("https://raw.githubusercontent.com/cbjuliana/datascience/master/quarentenadados/ml-latest-small/movies.csv")
filmes.columns = ["filmeId", "titulo", "generos"]
# traz as 5 primeiras linhas do arquivo
filmes.head()

#lendo a documentação de um método/atributo
?filmes.head()
```

```
avaliacoes = pd.read_csv("https://raw.githubusercontent.com/cbjuliana/datascience/master/quarentenadados/ml-latest-small/ratings.csv")

avaliacoes.head()
```

```
# número de linhas e colunas
avaliacoes.shape

len(avaliacoes)

avaliacoes.columns = ["usuarioId", "filmeId", "nota", "momento"]
avaliacoes.head()

avaliacoes.query("filmeId==1")

avaliacoes_do_filme_1 = avaliacoes.query("filmeId==1")
avaliacoes_do_filme_1.head()

#analisa as estatítiscas gerais dos dados
avaliacoes_do_filme_1.describe()

#repare abaixo como calculamos apenas a média das avaliações do filmeId 1.
avaliacoes_do_filme_1.mean()

# describe para a base completa
avaliacoes.describe()

avaliacoes["nota"]

#calculas a média
avaliacoes["nota"].mean()

avaliacoes_do_filme_1["nota"].mean()

#nota média para cada filme
notas_medias_por_filme = avaliacoes.groupby("filmeId")["nota"].mean()
notas_medias_por_filme.head()

filmes
notas_medias_por_filme

#join medias por filme
filmes_com_media = filmes.join(notas_medias_por_filme, on="filmeId")
filmes_com_media.head()

filmes_com_media.sort_values("nota", ascending=False).head(15)

#gerando gráfico
avaliacoes.query("filmeId == 1")["nota"].plot()

#escolhendo o tipo de gráfico
avaliacoes.query("filmeId == 2")["nota"].plot(kind='hist')

avaliacoes.query("filmeId == 1")["nota"].plot(kind='hist',
                                              title="Avaliações do filme Toy Story")


avaliacoes_filme_1 = avaliacoes.query("filmeId == 1")

notas_avaliacoes_filme_1 = avaliacoes_filme_1["nota"]

?notas_avaliacoes_filme_1.plot

#tipos de gráfico
    #- 'line' : line plot (default)
    #- 'bar' : vertical bar plot
    #- 'barh' : horizontal bar plot
    #- 'hist' : histogram
    #- 'box' : boxplot
    #- 'kde' : Kernel Density Estimation plot
    #- 'density' : same as 'kde'
    #- 'area' : area plot
    #- 'pie' : pie plot
    #- 'scatter' : scatter plot
    #- 'hexbin' : hexbin plot.

notas_avaliacoes_filme_1.plot(kind='line')


```
